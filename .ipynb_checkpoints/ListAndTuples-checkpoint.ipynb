# Tuples

* Tuples are an ordered sequence.
* Tuples example -> tuple1 = (5, 'hello', 10)
* concatination, Slicing, etc
* Tuples are immutable.
* tuples inside another tuple -> nesting.

# Lists

* Lists are also ordered sequences.
* List is represented by quare brackets -> list1 = ['rizwan', 12, 10.5]
* List are mutable. -> key difference from Tuples.
* Slicing, Concatination,
* List methods: list.append(), list.extend(), list.split(), aliasing,
* clone a list -> listA = [10, 15, 20] , listB = A[:]
* help command -> help(listA)

```python
list = ["The Bodyguard", 7.0, 1992, "BG", 1]
print(list[3:5])
```

```python
list.extend(['pop', 10])
print(list)
```

```python
list.append(['hello',10])
print(list)
```

```python
# Copy List

listA = ['hard rock', 10, 1.2]
listB = listA # listB and listA points to same reference
print('listA: ' , listA)
print('listB: ' , listB)
listA[0] = 2400
print('After modification')
print('listA: ' , listA)
print('listB: ' , listB)
```

```python
# Clone List

listC = ['heavy', 100]
listD = listC[:] # Now lisTD points to new memory location references 
print('listC: ' , listC)
print('listD: ' , listD)
listC[0] = 2400
print('After modification')
print('listC: ' , listC)
print('listD: ' , listD)
```
